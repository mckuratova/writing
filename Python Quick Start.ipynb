## Установка и настройка окружения для Windows

### Установка Python

1. Перейдите на [официальный сайт Python](https://www.python.org/downloads/windows/) и скачайте последнюю версию Python 3 (*версия 3.11.4 на 25.07.2023*).

    **Обратите внимание**, что данная версия не работает на Windows 7 и ниже.

2. Запустите скачанный exe-файл. Убедитесь, что установлены флажки:
    - Use admin privileges when installing py.exe
    - Add python.exe to PATH

3. Нажмите **Install Now**, чтобы выполнить установку с параметрами, заданными по умолчанию. Чтобы изменить эти параметры, выберите пункт **Customize installation** и следуйте инструкциям.

    После успешной установки переходите к следующему этапу.

### Создание виртуального окружения

При работе с Python удобно пользоваться виртуальным окружением, потому что оно помогает избежать конфликтов. Например, для одной программы можно установить одну версию библиотеки, а для другой — вторую.

Виртуальное окружение можно расположить где угодно, и директория тоже может называться как угодно. Однако распространенная практика — создавать виртуальное окружение рядом с кодом. Для примера создадим директорию *my_first_project* и сохраним в нее виртуальное окружение *env*.

1. Перейдите в директорию, где будет храниться новый проект:

    `cd c:\python_projects\my_first_project`



2. Создайте виртуальное окружение `env`:

    `python -m venv env`
    


Модуль [venv](https://docs.python.org/3/library/venv.html) создаст указанную директорию `env` и положит в нее служебные папки и файлы.

3. Теперь активируйте виртуальное окружение:

    `env\Scripts\activate.bat`
    


Обратите внимание, что вместо рабочего каталога теперь указан каталог с виртуальным окружением. Это означает, что вы внутри виртуального окружения. Теперь в него можно устанавливать любые пакеты.

4. Чтобы закончить работу с виртуальным окружением, закройте окно терминала или наберите команду `deactivate`.



### Настройка виртуального окружения

#### Обновление pip

В Python модули и библиотеки обычно устанавливаются с помощью менеджера пакетов `pip`. В Windows `pip` устанавливается через стандартный установщик Python. Чтобы убедиться в том, что вы используете последнюю версию `pip`, запустите следующую команду:

`python.exe -m pip install --upgrade pip`

В результате `pip` устанавливается в виртуальное окружение или обновляется до последней версии.

#### Установка библиотеки

Установим в виртуальное окружение `env` библиотеку для визуализации данных **Matplotlib**:

`pip install matplotlib`



Обратите внимание, что устанавливается сама библиотека *matplotlib*, а также зависимые библиотеки, среди которых *numpy*. Далее эти обе библиотеки пригодятся нам для создания первой программы на Python.

### Создание и запуск программы

#### Создание простой программы

Теперь в директории `my_first_project` создайте файл `main.py` и скопируйте в него следующий код:

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_it(a, b, c):
    x = np.linspace(-100, 100, 1000)
    y = a * x ** 2 + b * x + c

    fig, ax = plt.subplots()
    ax.plot(x, y)
    plt.show()

print('Квадратическая функция y=ax^2+bx+c. Введите коэффициенты:')
a = float(input('a = '))
b = float(input('b = '))
c = float(input('c = '))
draw_it(a, b, c)
```

В этой небольшой программе выполняется импорт библиотеки `numpy` и модуля `pyplot` библиотеки `matplotlib`. Определяется функция `draw_it`, которая принимает на вход коэффициенты $a$, $b$ и $c$ квадратичной функции $$f(x) = ax^2 + bx + c$$ и отображает график. Коэффициенты параболы вводит пользователь.

#### Запуск программы

В окне командной строки убедитесь, что виртуальное окружение `env` активно. Если нет, то активируйте его.  
Запустите программу:

`python main.py`



Программа отобразит заданный график:



## Установка и настройка среды разработки

Чтобы упростить написание, запуск и отладку программы, используют интегрированные среды разработки или IDE. Для Python их существует очень много. Рассмотрим подробнее одну из них: **PyCharm**. Установим *PyCharm* и создадим новый проект на основе ранее созданного примера для отображения параболы `main.py` и виртуального окружения `env`.

### Установка PyCharm

Перейдите на официальный сайт https://www.jetbrains.com/ru-ru/pycharm/download/?section=windows.

Прочтите системные требования для текущей версии программы. Убедитесь, что ваша система им соответствует.

Скачайте бесплатную версию Community Edition:

Запустите скачанный файл и следуйте инструкциям мастера установки.

### Настройка PyCharm

Запустите программу PyCharm. В появившемся окне слева выберите вкладку *Customize (Настроить)*. Здесь можно выбрать цвет темы (*Color theme*) и размер шрифта (*Accessibility*), а также настроить сочетания клавиш (*Keymap*):

## Создание первого проекта

1. Чтобы создать проект, перейдите на вкладку *Projects (Проекты)* и нажмите на значок + *New Project (Новый проект)*.

2. В появившемся окне укажите местоположение проекта `my_first_project`, выберите пункт *Previously configured interpreter (Ранее настроенный интерпретатор)* и в выпадающем меню *Add Interpreter (Добавить интерпретатор)* выберите пункт *Add Local Interpreter... (Добавить локальный интерпретатор)*.

3. Выберите *Existing (Существующее)* окружение и в поле *Interpreter (Интерпретатор)* укажите путь до интерпретатора в ранее созданном виртуальном окружении `env`: `...\my_first_project\env\Scripts\python.exe`



4. Нажмите *OK*. Теперь интерпретатор из виртуального окружения `env` будет отображаться в окне создания проекта.
    


5. Нажмите *Create*. Появится сообщение о том, что директория не пустая (*Directory Is Not Empty*). Нажмите *Create from Existing Sources*, чтобы создать проект на основе имеющихся в директории файлов (*в нашем случае это созданный в пункте 1.4.1 файл main.py*).



6. Нажмите на `main.py` и откроется файл, созданный в пункте 1.4.1.



## Запуск программы в отладке

### Запуск программы

Чтобы запустить программу: 
- Нажмите на зеленый треугольник на панели инструментов вверху **ИЛИ**
- Нажмите на значок *Меню* ≣ → *Run* → *Run 'main.py'* **ИЛИ**
- Нажмите Shift+F10.



В нижней части окна появится поле ввода коэффициентов. Введите коэффициенты и нажмите ENTER. График функции отобразится в новом окне.



### Отладка программы

Чтобы найти ошибки и проверить работу программы, используется режим отладки. Для отладки установите точку останова, то есть выберите строку программы, начиная с которой будете отслеживать ее выполнение.

1. Чтобы установить точку останова:
- Наведите мышкой на номер строки и нажмите на него ИЛИ
- Выделите нужную строку кода и выберите *Меню* ≣ → *Run* → *Toggle Breakpoint* → *Line Breakpoint* ИЛИ
- Выделите нужную строку кода и нажмите CTRL+F8.

    Напротив строки будет установлена красная точка:



2. Чтобы запустить отладку:
- Нажмите на значок жучка на панели инструментов вверху **ИЛИ**
- Выберите *Меню* ≣ → *Run* → *Debug 'main.py'* **ИЛИ**
- Нажмите Shift+F9.



3. После ввода коэффициентов выполнение программы остановится на входе в функцию `draw_it`. На нижней панели есть несколько кнопок для управления программой.  
- Кнопка 1: не входить в функцию;
- Кнопка 2: войти в функцию;
- Кнопка 3: возобновить работу программы.



Текущее положение интерпретатора отмечается синей строкой. Нажмите на кнопку 2 *Step Into* и увидите, как интерпретатор перешел в функцию `draw_it`. Чтобы выйти из функции, нажмите кнопку 1 *Step Out*. Чтобы продолжить выполнение программы без отладки, нажмите кнопку 3 *Resume Program*.

Чтобы прекратить отладку и выполнение программы:
- Нажмите на красный квадрат на панели инструментов вверху окна **ИЛИ**
- Выберите *Меню* ≣ → *Run* → *Stop 'main.py'* **ИЛИ**
- Нажмите CTRL+F2.

Таким образом вы легко проследите, что происходит в вашей программе.
